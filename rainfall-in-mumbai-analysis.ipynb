# INSIGHTS AND VISUALIZATIONS 📊

```python
data.nunique()
```

```python
# Calculate annual averages and other statistics
annual_avg = data['Total'].mean()
annual_max = data['Total'].max()
annual_min = data['Total'].min()

print(f"Annual Average Rainfall: {annual_avg:.2f} mm")
print(f"Maximum Annual Rainfall: {annual_max:.2f} mm")
print(f"Minimum Annual Rainfall: {annual_min:.2f} mm")
```

```python
# Plotting Total Rainfall Over the Years
plt.figure(figsize=(18, 8))
plt.plot(data['Year'], data['Total'], marker='o')
plt.title('Total Annual Rainfall in Mumbai (1901-2021)')
plt.xlabel('Year')
plt.ylabel('Total Rainfall (mm)')
plt.grid()
plt.xticks(data['Year'], rotation=90)
plt.show()
```

```python
# Monthly Rainfall Analysis
monthly_avg = data.iloc[:, 1:-1].mean()
plt.figure(figsize=(10, 6))
monthly_avg.plot(kind='bar')
plt.title('Average Monthly Rainfall in Mumbai (1901-2021)')
plt.xlabel('Month')
plt.ylabel('Average Rainfall (mm)')
plt.xticks(rotation=90)
plt.show()
```

```python
# Box plot for monthly rainfall distribution
monthly_data = data.iloc[:, 1:-1]
plt.figure(figsize=(12, 6))
monthly_data.boxplot()
plt.title('Monthly Rainfall Distribution in Mumbai (1901-2021)')
plt.ylabel('Rainfall (mm)')
plt.xlabel('Month')
plt.xticks(rotation=90)
plt.grid()
plt.show()
```

```python
# Cumulative rainfall
data['Cumulative'] = data['Total'].cumsum()
plt.figure(figsize=(18, 8))
plt.plot(data['Year'], data['Cumulative'], marker='o', color='orange')
plt.title('Cumulative Annual Rainfall in Mumbai (1901-2021)')
plt.xlabel('Year')
plt.ylabel('Cumulative Rainfall (mm)')
plt.grid()
plt.xticks(data['Year'], rotation=90)
plt.show()
```

```python
plt.figure(figsize=(12, 6))
plt.hist(data['Total'], bins=20, color='skyblue', edgecolor='black')
plt.title('Distribution of Annual Rainfall in Mumbai (1901-2021)')
plt.xlabel('Total Rainfall (mm)')
plt.ylabel('Frequency')
plt.show()
```

```python
# A bar chart comparing average rainfall for different seasons
seasons = {
    'Pre-Monsoon': data[['Jan', 'Feb', 'Mar', 'April', 'May']].sum(axis=1),
    'Monsoon': data[['June', 'July', 'Aug', 'Sept']].sum(axis=1),
    'Post-Monsoon': data[['Oct', 'Nov', 'Dec']].sum(axis=1)
}
seasonal_data = pd.DataFrame(seasons)
# Bar chart
seasonal_means = seasonal_data.mean()
plt.figure(figsize=(10, 4))
seasonal_means.plot(kind='bar', color=['lightcoral', 'skyblue', 'lightgreen'])
plt.title('Average Seasonal Rainfall in Mumbai (1901-2021)')
plt.ylabel('Average Rainfall (mm)')
plt.xlabel('Season')
plt.xticks(rotation= 0)
plt.show()
```
